### 1. Which establishments have a hygiene score equal to 20?

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python

# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
# Define the query to find establishments in London with a RatingValue greater than or equal to 4
query2 = {
    "LocalAuthorityName": {"$regex": ".*London.*", "$options": "i"},
    "RatingValue": {"$gte": "4"}
}

# Find the establishments matching the query
results = db.establishments.find(query2)

# Count the number of documents in the result
num_documents = db.establishments.count_documents(query2)
print("Number of documents:", num_documents)

# Check if there are any matching documents
if num_documents > 0:
    # Print the first document in the results using pprint
    pprint(list(results)[0])
else:
    print("No documents found.")
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Define the aggregation pipeline
pipeline = [
    {"$match": {"scores.Hygiene": 0}},  # Match establishments with hygiene score 0
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}}, # Group by Local Authority and count the establishments
    {"$sort": {"count": -1}}, # Sort the results by count in descending order
    {"$limit": 10} # Limit the results to top 10
]

# Perform the aggregation query
results = establishments.aggregate(pipeline)

# Initialize a counter for counting the documents
count_documents = 0

# Iterate over the results and print each document
for result in results:
    print(result)
    count_documents += 1

# Print the total number of documents
print("Number of documents:", count_documents)

```

```python
# Convert the result to a Pandas DataFrame
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},
    {"$limit": 10}
]

results = establishments.aggregate(pipeline)

# Convert the result to a list of dictionaries
result_list = list(results)

# Convert the list of dictionaries to a Pandas DataFrame
df = pd.DataFrame(result_list)

# Display the number of rows in the DataFrame
num_rows = len(df)
print("Number of rows:", num_rows)

# Display the first 10 rows of the DataFrame
df.head(10)
```
