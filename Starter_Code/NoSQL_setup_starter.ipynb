1. An exciting new halal restaurant just opened in Greenwich, but hasn't been rated yet. The magazine has asked you to include it in your analysis. Add the following restaurant "Penang Flavours" to the database.

```python
# Create a dictionary for the new restaurant data
# Create a new document for the halal restaurant
new_restaurant = {
    "BusinessName": "Penang Flavours",
    "BusinessType": "Restaurant/Cafe/Canteen",
    "BusinessTypeID": "",
    "AddressLine1": "Penang Flavours",
    "AddressLine2": "146A Plumstead Rd",
    "AddressLine3": "London",
    "AddressLine4": "",
    "PostCode": "SE18 7DY",
    "Phone": "",
    "LocalAuthorityCode": "511",
    "LocalAuthorityName": "Greenwich",
    "LocalAuthorityWebSite": "http://www.royalgreenwich.gov.uk",
    "LocalAuthorityEmailAddress": "health@royalgreenwich.gov.uk",
    "scores": {
        "Hygiene": "",
        "Structural": "",
        "ConfidenceInManagement": ""
    },
    "SchemeType": "FHRS",
    "geocode": {
        "longitude": "0.08384000",
        "latitude": "51.49014200"
    },
    "RightToReply": "",
    "Distance": 4623.9723280747176,
    "NewRatingPending": True
}


```

2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen"
result = establishments.find_one(
    {"BusinessType": "Restaurant/Cafe/Canteen"},
    {"BusinessTypeID": 1, "BusinessType": 1}
)

# Print the result
if result:
    pprint(result)
else:
    print("Business type not found.")
```

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
# Define the query to find the BusinessTypeID
query = {"BusinessType": "Restaurant/Cafe/Canteen"}

# Define the projection to include only the BusinessTypeID and BusinessType fields
projection = {"BusinessTypeID": 1, "BusinessType": 1, "_id": 0}

# Find the matching documents and project only the specified fields
result = establishments.find(query, projection)

# Iterate over the result and print each document
for document in result:
    pprint(document)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
# Define the query to find the new restaurant by BusinessName
query = {"BusinessName": "Penang Flavours"}

# Define the updated BusinessTypeID
new_business_type_id = 1

# Define the update operation
update_operation = {"$set": {"BusinessTypeID": new_business_type_id}}

# Update the document with the new BusinessTypeID
result = establishments.update_one(query, update_operation)

# Check if the update was successful
if result.modified_count > 0:
    print("New restaurant updated successfully.")
else:
    print("New restaurant update failed.")
```

```python
# Confirm that the new restaurant was updated
# Define the query to find the updated restaurant by BusinessName
query = {"BusinessName": "Penang Flavours"}

# Find the updated restaurant document
updated_restaurant = establishments.find_one(query)

# Print the updated restaurant document
if updated_restaurant:
    pprint(updated_restaurant)
else:
    print("Updated restaurant not found.")
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
# Define the query to find documents with LocalAuthorityName as "Dover"
query = {"LocalAuthorityName": "Dover"}

# Count the number of documents matching the query
count = establishments.count_documents(query)

# Print the count
print("Number of documents with LocalAuthorityName as 'Dover':", count)
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
# Define the query to find documents with LocalAuthorityName as "Dover"
query = {"LocalAuthorityName": "Dover"}

# Delete the matching documents
result = establishments.delete_many(query)

# Check if the deletion was successful
if result.deleted_count > 0:
    print("Documents with LocalAuthorityName as 'Dover' deleted successfully.")
else:
    print("No documents with LocalAuthorityName as 'Dover' found.")

```

```python
# Check if any remaining documents include Dover
# Define the query to find documents with LocalAuthorityName including "Dover"
query = {"LocalAuthorityName": {"$regex": "Dover"}}

# Count the number of documents matching the query
count = establishments.count_documents(query)

# Print the count
print("Number of documents with LocalAuthorityName including 'Dover':", count)
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
# Import the Decimal module
from decimal import Decimal

# Define the update operation
update_operation = [
    {"$set": {"longitude": {"$toDouble": "$longitude"}}},
    {"$set": {"latitude": {"$toDouble": "$latitude"}}}
]

# Update the documents to convert longitude and latitude to decimal numbers
result = establishments.update_many({}, update_operation)

# Check if the update was successful
if result.modified_count > 0:
    print("Longitude and latitude data type conversion completed successfully.")
else:
    print("No documents were updated.")

```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
# non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
# establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
# Define the non-rating values to set as null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]

# Update the documents to convert RatingValue to integers and set non-rating values as null
result = establishments.update_many(
    {"RatingValue": {"$nin": [1, 2, 3, 4, 5]}},
    {"$set": {"RatingValue": None}}
)

# Check if the update was successful
if result.modified_count > 0:
    print("RatingValue conversion and update completed successfully.")
else:
    print("No documents were updated.")
```

```python
# Change the data type from String to Integer for RatingValue
# Define the update operation
update_operation = {"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}

# Update the documents to convert RatingValue to integer
result = establishments.update_many({}, update_operation)

# Check if the update was successful
if result.modified_count > 0:
    print("RatingValue data type conversion completed successfully.")
else:
    print("No documents were updated.")
```
